```python
#importing needed libraries 
%matplotlib inline 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
#fetch the dataset to be analysed

from sklearn.datasets import fetch_20newsgroups
data = fetch_20newsgroups()
data.target_names
```

```python
#define the categories
categories =['alt.atheism',
 'comp.graphics',
 'comp.os.ms-windows.misc',
 'comp.sys.ibm.pc.hardware',
 'comp.sys.mac.hardware',
 'comp.windows.x',
 'misc.forsale',
 'rec.autos',
 'rec.motorcycles',
 'rec.sport.baseball',
 'rec.sport.hockey',
 'sci.crypt',
 'sci.electronics',
 'sci.med',
 'sci.space',
 'soc.religion.christian',
 'talk.politics.guns',
 'talk.politics.mideast',
 'talk.politics.misc',
 'talk.religion.misc']
```

```python
#train data on these categories

train = fetch_20newsgroups(subset='train', categories=categories)

#testing the data on these categories 

test = fetch_20newsgroups(subset='test', categories=categories)
print(train.data[5])
```

```python
##let us see the lenght of the number of news articles we have 

print(len(train.data))
```

```python
#importing the nb algth libraries 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline 

#create a model based on MNB
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

#train the model we created above 

model.fit(train.data, train.target)

#creating labels for the test data

labels = model.predict(test.data)
```

```python
#create our confusion matrix 
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, 
            xticklabels=train.target_names,
            yticklabels=train.target_names)
#plotting heat maps

plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
#predict a new category on the new data we are striving for based on the model we trained. Building a user 

def predict_category(s, train=train, model=model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
##run our predict model to see how well its performing 
```
